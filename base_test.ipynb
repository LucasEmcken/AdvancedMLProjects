```python
#Data loading and splitting
device = 'cpu'
dataset = TUDataset(root='./data/', name='MUTAG').to(device)
rng = torch.Generator().manual_seed(0)
train_dataset, validation_dataset, test_dataset = random_split(dataset, (100, 44, 44), generator=rng)
train_loader = DataLoader(train_dataset, batch_size=100)
validation_loader = DataLoader(validation_dataset, batch_size=44)
test_loader = DataLoader(test_dataset, batch_size=44)
```

```python
train_graph_sizes = []
for data in train_dataset:
    train_graph_sizes.append(data.num_nodes)

def get_N():
    return random.choice(train_graph_sizes)

def possible_edges_func(num_nodes):
    return num_nodes * (num_nodes - 1) // 2

def link_probability(num_edges,num_nodes):
    r = num_edges/possible_edges_func(num_nodes)
    return r

#Erdos-Renyi graph generation
def generate_erdos_renyi_graph(num_nodes, p):
    return erdos_renyi_graph(num_nodes, p, directed=False)

def generate_erdos_renyi_graphs(num_graphs, num_nodes, p):
    graphs = []
    for _ in range(num_graphs):
        graph = generate_erdos_renyi_graph(num_nodes, p)
        graphs.append(graph)
    return graphs
nn = get_N()
graphs = generate_erdos_renyi_graphs(10,get_N(),link_probability(possible_edges_func(nn),nn))
graphs
```
